```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

# Sample DataFrame
answer_path = "scores.csv"
df = pd.read_csv(answer_path) 

# Count the frequency of values in column 'abc' and 'xyz'
abc_counts = df['score'].value_counts().sort_index()
# xyz_counts = df['finetunedModeAnswerScore'].value_counts().sort_index()

# Define colors for each bar (for both graphs)
colors_abc = ['blue', 'orange', 'green', 'red', 'purple', 'brown']
# colors_xyz = ['blue', 'orange', 'green', 'red', 'purple', 'brown']

# Create subplots
fig, axs = plt.subplots(1, 1, figsize=(6, 6))

# Plotting the first bar graph for column 'abc'
axs.bar(abc_counts.index, abc_counts.values, color=colors_abc)
axs.set_xlabel('Categories')
axs.set_ylabel('Frequency')
axs.set_title('Mistral 7B rag evaluation')
axs.xaxis.set_major_locator(MaxNLocator(integer=True))  # Ensure integer ticks on x-axis

# # Plotting the second bar graph for column 'xyz'
# axs[1].bar(xyz_counts.index, xyz_counts.values, color=colors_xyz)
# axs[1].set_xlabel('Categories')
# axs[1].set_ylabel('Frequency')
# axs[1].set_title('FineTuned Model')
# axs[1].xaxis.set_major_locator(MaxNLocator(integer=True))  # Ensure integer ticks on x-axis

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
